```python
corn = glob('./corn/train/*.png')
```

```python
img_mpl = plt.imread(corn[10])
img_cv2 = cv2.imread(corn[20])
```

```python
print(img_mpl.shape)
print(img_cv2.shape)
# interesting we don't have images which are all the same size 
```

```python
print(type(corn))
print(corn)

shapes = [plt.imread(img).shape for img in corn[:30]]
print(shapes) # lots of different shapes 
print(type(shapes), type(shapes[1]))
```

```python
# Get unique shapes: 
uni_shapes = list(set(shapes))
print(len(uni_shapes)) 
print(uni_shapes)

# get the range of possible values 
position_values = list(zip(*uni_shapes)) # supplying a bunch of tuples ... * strips the list 
print("min:", [min(values) for values in position_values])
print("max:", [max(values) for values in position_values])
print("mean:", [np.mean(values) for values in position_values])

# If I wanted to learn more about the shapes of these images I could create histograms for each profile or each dimension to see what the images average is. 
# Don't think this is necessary for now, just worth noting that there is some irregularity to the data and a pretty large range of different values / resolutions
```

```python
print(type(img_cv2), type(img_mpl))
print(img_mpl)
print(np.max(img_mpl), np.min(img_mpl), np.mean(img_mpl)) # looking at the range of pixel values of a single image 
```

## Image Manipulation 


```python
img_gray = cv2.cvtColor(plt.imread(corn[4]), cv2.COLOR_RGB2GRAY)
fig, ax = plt.subplots(figsize=(5,5))
ax.imshow(img_gray, cmap='Greys') # different grey spelling 
ax.set_title('greyscale version of an image')
```

```python
img_resize = cv2.resize(img_gray, None, fx=1.5, fy=1.5)
fig, ax = plt.subplots(figsize=(5,5))
ax.imshow(img_resize, cmap='Greys') # different grey spelling 
ax.set_title('greyscale version of an image that is resized ')
# seems like some default interpolation was decided on.. 
```

```python
img_resize = cv2.resize(img_gray,None, fx=2.5, fy=2.5, interpolation = cv2.INTER_CUBIC)
fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].imshow(img_resize, cmap='Greys') # different grey spelling 
ax[1].imshow(plt.imread(corn[4])) 
# ax.set_title('greyscale version of an image that is resized ')  # <- setting title here actually throws an error for some reason 
# seems like some default interpolation was decided on.. 
print(img_resize.shape, img_gray.shape) # we can see there are more values used to represent the image now. I think most of the original data has been preserved for this image 
```
